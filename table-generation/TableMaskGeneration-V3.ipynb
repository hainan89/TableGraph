## 表格图像语义分割标注图像

```python
import pandas as pd
import numpy as np
import cv2
import os
```

```python
dataset = './dataset04/'
imgs_fold = dataset + "imgs/"
label_fold = dataset + "labels/"
mask_fold = dataset + "/masks/"
```

```python
label_fns = [f for f in os.listdir(label_fold)]

for one in label_fns:
    
    # columns: l_x, l_y, r_x, r_y, s_x, s_y, w, h, unit_txt
    labels = pd.read_csv(os.path.join(label_fold, one), index_col=0, header=0)
    
    img_n = one.replace('.label', '')
    img_fpath = os.path.join(imgs_fold, img_n)
#     print(img_fpath)
    
    image = cv2.imread(img_fpath)
    mask_img = np.array(np.ones(image.shape) * 255, dtype="uint8")
    for i in np.arange(labels.shape[0]):
        l_x = labels.iloc[i, 0]
        l_y = labels.iloc[i, 1]
        r_x = labels.iloc[i, 2]
        r_y = labels.iloc[i, 3]
        line_w = labels.iloc[i, 4]
        s_x = labels.iloc[i, 5]
        s_y = labels.iloc[i, 6]
        s_w = labels.iloc[i, 7]
        s_h = labels.iloc[i, 8]
#         color = [np.random.randint(0, 256), np.random.randint(0, 256), np.random.randint(0, 256)]
        
        
        ## wide cross rectangle mask
#         cv2.rectangle(mask_img, (int((r_x - l_x) / 3 + l_x), l_y), (int((r_x - l_x) * 2 / 3 + l_x), r_y), (255, 255, 255), -1)
#         cv2.rectangle(mask_img, (l_x, int((r_y - l_y) / 3 + l_y)), (r_x, int((r_y - l_y) * 2 / 3 + l_y)), (255, 255, 255), -1)
        
        # poly mask
#         cv2.fillConvexPoly(mask_img, np.array([[int((r_x - l_x) / 2 + l_x), l_y],
#                                 [r_x, int((r_y - l_y) / 2 + l_y)],
#                                 [int((r_x - l_x) / 2) + l_x, r_y],
#                                 [l_x, int((r_y - l_y) / 2 + l_y)]]), (255, 255, 255))
        
        
        # mask for the cell areas
#         cv2.rectangle(mask_img, (l_x+5,l_y+5), (r_x-5, r_y-5), (0, 0, 255), -1)
        
        # mask for the unit texts
        cv2.rectangle(mask_img, (s_x, s_y), (s_x + s_w, s_y + s_h), (125, 125, 125), -1)
        
        # mask for the cell lines
        if l_x > 0 and l_y > 0 and r_x > 0 and r_y > 0 and line_w > 0:
            cv2.rectangle(mask_img, (l_x,l_y), (r_x, r_y), (0, 0, 0), line_w)
    
    if not os.path.exists(mask_fold):
        os.mkdir(mask_fold)
    mask_fpath = os.path.join(mask_fold, img_n)
    
    mask_img_u = cv2.cvtColor(mask_img, cv2.COLOR_BGR2GRAY)
    
    cv2.imwrite(mask_fpath, mask_img_u)
    
    print(one)
#     break
```
